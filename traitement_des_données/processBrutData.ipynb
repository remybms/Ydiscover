```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
df = pd.read_csv('./nantes_test.csv')
```

```python
df
```

```python
df= df.drop(columns=['severerisk', 'stations'])
```

```python
mapping = {np.nan: 0, 'rain' : 1, 'rain,snow' : 2}
```

```python
df = df.replace({'preciptype' : mapping})
df
```

```python
def fToCDegres(x):
    x = (x-32)*5/9
    return float(x)

def oneK(x):
    return x*1000
```

```python
df['temp'] = df['temp'].apply(fToCDegres)
df['feelslike'] = df['feelslike'].apply(fToCDegres)
df['dew'] = df['dew'].apply(fToCDegres)
df['precip'] = df['precip'].apply(oneK)
pd.set_option('display.precision', 1)
df
```

```python
df.to_csv(path_or_buf="./nantesPrepared.csv")
```
